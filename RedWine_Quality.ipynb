# Feature Selection

```python
# Create Classification version of target variable
# redwine_df['goodquality'] = [1 if x > 6.5 else 0 for x in redwine_df['quality']]# Separate feature variables and target variable
# X = redwine_df.drop(['quality','goodquality'], axis = 1)
# Y = redwine_df['goodquality']
```

```python
redwine_df["quality"] = np.where(redwine_df["quality"] >=6 , 1, 0) 
```

```python
redwine_df.head()
```

```python
X=redwine_df.values[:,0:-1]
Y=redwine_df.values[:,-1]
```

```python
print(X.shape, Y.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
X=scaler.fit_transform(X)#fit will lear mean and sigma values 
#scaler.fit_transform(X)
#X=scaler.transform(X)
print(X)
```

# Splitting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.3,random_state=10)
```

```python
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train,Y_train)
Y_pred = model.predict(X_test)

from sklearn.metrics import accuracy_score,confusion_matrix
print("Accuracy Score:",accuracy_score(Y_test,Y_pred))
```

```python
confusion_mat = confusion_matrix(Y_test,Y_pred)
print(confusion_mat)
```

```python
# store the predicted probabilities
y_pred_prob=model.predict_proba(X_test)
print(y_pred_prob)
```

```python
#Evaluation matrix

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```
